```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
validation_loss_file_paths = [
    r'D:\Praharsha\code\CAMZ\models\model_history\1.0-CNN\CNN_loss_logger_val_batch_wise.csv',
    r'D:\Praharsha\code\CAMZ\models\model_history\2.0-VGG16\loss_logger_val_batch_wise.csv',
    r'D:\Praharsha\code\CAMZ\models\model_history\3.0-DenseNet121\loss_logger_val_batch_wise.csv',
    # r'D:\Praharsha\code\CAMZ\models\model_history\4.0-MobileNetv2\loss_logger_val_batch_wise.csv'
    # r'D:\Praharsha\code\CAMZ\models\model_history\5.0-ResNet\loss_logger_val_batch_wise.csv',
    # r'D:\Praharsha\code\CAMZ\models\model_history\6.0-EfficientNetv2\loss_logger_val_batch_wise.csv',
    # r'D:\Praharsha\code\CAMZ\models\model_history\7.0-Inceptionv3\loss_logger_val_batch_wise.csv'
]

validation_df_list = [
    pd.read_csv(file_path) for file_path in validation_loss_file_paths
]
```

```python
validation_batch_loss_list = [
    df['loss'].values for df in validation_df_list
]

validation_epoch_loss_list = [
    np.array([np.mean(df[df['epoch'] == epoch]['loss'].values) for epoch in range(df['epoch'].min(), df['epoch'].max()+1)]) for df in validation_df_list
]
```

```python
fig = plt.figure(figsize=(10, 6))

num_models = 3
for i in range(num_models):
    plt.plot(validation_batch_loss_list[i],alpha = 0.7)
    plt.plot(np.linspace(0,len(validation_batch_loss_list[i]),len(validation_epoch_loss_list[i])),validation_epoch_loss_list[i])
```
